```python
dados['Pontos_extras'] = dados['Notas'].apply(lambda x: x*0.4)
dados
```

```python
dados['Notas_finais'] = dados['Notas'] + dados['Pontos_extras']
dados
```

```python
dados['Aprovado_final'] = dados['Notas_finais'].apply(lambda x:True if x >= 6 else False)
dados
```
